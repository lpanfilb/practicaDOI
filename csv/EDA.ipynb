Después de la exploración, las columnas elegidas para este dataset son:

- `Ceremony` -(renombrado)-> `Event`
- `Class` -(renombrado)-> `Category`
- `CanonicalCategory` -(suplantado)->`award` (segundo dataset)
- `MultfilmNomination` 7 (DataType Person)
- `Year` 7 (Datatype Ceremony)
- `Winner` 7 (Datatype Person)
- `Detail` -(renombrado)-> `Character`

Algunas aclaraciones sobre las columnas:

- `Class` $\to$ Es una categoría general (macro). Agrupa premios en bloques amplios.
- `CanonicalCategory` $\to$ Es la categoría específica del premio (el nombre “oficial” o estandarizado).
- `MultifilmNomination` $\to$ Anteriormente un actor podía estar nominado por varias películas y no solo uno como actualmente.
